# NHANES dataset

This Python notebook will read some files of the NHANES dataset.<br>

The NHANES dataset is an American open-source dataset which stands for the National Health And Nutrition Examination Survey.

All data files are stored on the Centers and Disease Control website: https://wwwn.cdc.gov/nchs/nhanes/Default.aspx.<br>
There are data files dating from the year 1999-2026.<br>
The data files are stored in data files from a 2-year time frame.

Each 2-year time frame contains a lot of different variables of almost 10.000 people.<br>
The most important datasets for our problem would be the demographics, blood pressure, glucose and cholesterol.<br>
These datasets can be found at the following:
- Demographics Data &rarr; Demographic Variables & Sample Weights
- Examination Data &rarr; Blood pressure
- Laboratory Data &rarr; Plasma Fasting Glucose & Insulin
- Laboratory Data &rarr; Cholesterol - HDL
- Laboratory Data &rarr; Cholesterol - Total

Each dataset has a representive documentation. This documentation is provided with an explanation about the different variables (column names), and shows us how many datapoints in the dataset belong to each categorical or numerical value, and how many rows are missing.<br>

The data files are saved as *.xpt files, they can be downloaded and read with this script.

## Data files

We first need to read the *.xpt files.<br>
We can read the *.xpt files with pandas and save the CSV files.

The first column of each dataset is the participant ID, so all datasets can be merged with each other.

The notebook below shows the most important variables of each dataset and a preview of how the dataset looks like.

```python
import pandas as pd
```

### Demographics dataset

- SEQN - Participant ID
- RIAGENDR - Gender (1=male; 2=female)
- RIDAGEYR - Age
- RIDRETH3 - Ethnicity (1=Mexican American, 2=Other Hispanic, 3=Non-Hispanic White, 4=Non-Hispanic Black, 6=Non-Hispanic Asian, 7=Other Race - Including Multi-Racial)

```python
# Read the NHANES demographics dataset from the XPT file
df_demographics = pd.read_sas("2017-2018/xpt/DEMO_J.XPT", format="xport")
# Save the DataFrame to a readable CSV file
df_demographics.to_csv("2017-2018/csv/DEMO_J.csv", index=False)
df_demographics
```

### Blood pressure dataset

- SEQN - Participant ID
- BPXCHR - Heart rate
- BPXPLS - Heart pulse
- BPXSY1 - Systolic blood pressure (top number)
- BPXDI1 - Diastolic blood pressure (bottom number)

Blood pressure is measured 4 times.

```python
# Read the NHANES blood pressure dataset from the XPT file
df_blood_pressure = pd.read_sas("2017-2018/xpt/BPX_J.XPT", format="xport")
# Save the DataFrame to a readable CSV file
df_blood_pressure.to_csv("2017-2018/csv/BPX_J.csv", index=False)
df_blood_pressure
```

### Glucose dataset

- SEQN - Participant ID
- LBXGLU - Fasting Glucose (mg/dL)
- LBDGLUSI - Fasting Glucose (mmol/L)

```python
# Read the NHANES glucose dataset from the XPT file
df_glucose = pd.read_sas("2017-2018/xpt/GLU_J.XPT", format="xport")
# Save the DataFrame to a readable CSV file
df_glucose.to_csv("2017-2018/csv/GLU_J.csv", index=False)
df_glucose
```

### Total Cholesterol dataset

- SEQN - Participant ID
- LBXTC - Total Cholesterol (mg/dL)
- LBDTCSI - Total Cholesterol (mmol/L)

```python
# Read the NHANES total cholesterol dataset from the XPT file
df_cholesterol = pd.read_sas("2017-2018/xpt/TCHOL_J.XPT", format="xport")
# Save the DataFrame to a readable CSV file
df_cholesterol.to_csv("2017-2018/csv/TCHOL_J.csv", index=False)
df_cholesterol
```

### HDL Cholesterol dataset

- SEQN - Participant ID
- LBDHDD - HDL Cholesterol (mg/dL)
- LBHDDSI - HDL Cholesterol (mmol/L)

```python
# Read the NHANES HDL cholesterol dataset from the XPT file
df_hdl = pd.read_sas("2017-2018/xpt/HDL_J.XPT", format="xport")
# Save the DataFrame to a readable CSV file
df_hdl.to_csv("2017-2018/csv/HDL_J.csv", index=False)
df_hdl
```
